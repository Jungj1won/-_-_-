```python
# 1. 최대 강우강도 지표 생성
rain['datetime'] = pd.to_datetime(rain[['year', 'month', 'day', 'hour']])

# 한 시간 내 최대 강우량 계산
rainfall_hour = rain.groupby(['district', rain['datetime'].dt.year, rain['datetime'].dt.month, rain['datetime'].dt.day, rain['datetime'].dt.hour])['rain(mm)'].max()

# 5년간 각 지역의 최대 강우량 계산
max_rainfall_hour = rainfall_hour.groupby('district').max()

# 인덱스 재설정
max_rainfall_hour = max_rainfall_hour.reset_index()

max_rainfall_hour.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 엘보우 방법을 사용하여 최적의 K값 탐색
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(exp)
    sse.append(kmeans.inertia_)

# SSE 그래프 그리기
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# 실루엣 계수을 사용하여 최적의 K값 탐색
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(exp)
    score = silhouette_score(exp, kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(range(2, 11), silhouette_scores, marker='o')
```

```python
from sklearn.cluster import KMeans

# 군집 개수
k = 3

# 그룹 수, random_state 설정
model = KMeans(n_clusters = k, random_state = 0)

# 정규화된 데이터에 학습
model.fit(exp)

# 클러스터링 결과 각 데이터가 몇 번째 그룹에 속하는지 저장
exp['cluster'] = model.fit_predict(exp)

exp_new = exp.copy()

exp_new
```

```python
# 군집화 결과 시각화
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# PCA를 통해 데이터를 2차원으로 축소
pca = PCA(n_components=2)
final_pca = pca.fit_transform(exp.drop('cluster', axis=1))

# PCA 결과와 클러스터 라벨을 DataFrame으로 변환
pca_df = pd.DataFrame(data=final_pca, columns=['PC1', 'PC2'])
pca_df['Cluster'] = exp['cluster']

# 각 클러스터별로 다른 색상으로 데이터 포인트를 시각화
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue', 'purple']

for i in range(3):
    plt.scatter(pca_df.loc[pca_df['Cluster'] == i, 'PC1'],
                pca_df.loc[pca_df['Cluster'] == i, 'PC2'],
                s=50, c=colors[i], label=f'Cluster {i}')

plt.title('2D PCA Visualization of KMeans Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
